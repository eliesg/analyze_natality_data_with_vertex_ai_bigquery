<p>In this notebook, we will take a look at a dataset of The Office episodes, and try to understand how the popularity and quality of the series varied over time. <p>This dataset contains information on a variety of characteristics of each episode. In detail, these are:

<div style="background-color: #efebe4; color: #05192d; text-align:left; vertical-align: middle; padding: 15px 25px 15px 25px; line-height: 1.6;">
    <div style="font-size:20px"><b>datasets/office_episodes.csv</b></div>

    <li><b>season:</b> Season in which the episode appeared.</li>

    <li><b>writers:</b> Writers of the episode.</li>
    <li><b>has_guests:</b> True/False column for whether the episode contained guest stars.</li>
    <li><b>scaled_ratings:</b> The ratings scaled from 0 (worst-reviewed) to 1 (best-reviewed).</li>

```python
# Use this cell to begin your analysis, and add as many as you would like!

import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [11, 7]

# create dataframe and preview the data
df = pd.read_csv("datasets/office_episodes.csv")
print(df.head())
```

```python
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]

# create color scheme
colors = []

for index, row in df.iterrows() :
    if row["scaled_ratings"] < 0.25 :
        colors.append("red")
    elif row["scaled_ratings"] < 0.50 :
        colors.append("orange")
    elif row["scaled_ratings"] < 0.75 :
        colors.append("lightgreen")
    else :
        colors.append("darkgreen")

# create scatter plot
plt.scatter(
    x = df.episode_number,
    y = df.viewership_mil,
    s = 25 + df.has_guests * 225,
    c = colors
)
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```
